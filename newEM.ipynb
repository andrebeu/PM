```python
from glob import glob as glob
import torch as tr
import numpy as np

from scipy.stats import pearsonr
from scipy.spatial import distance

from PMmodel import *

from matplotlib import pyplot as plt

%matplotlib inline

%load_ext autoreload
%reload_ext autoreload
%autoreload 2

plt.rcParams.update({'font.size':22})
```

```python
def eval_(net,task,neps=500,pmL=[5,9]):
  teseqlen = 15
  score = -np.ones([neps,teseqlen])
  for ep in range(neps):
    # generate data
    x_seq,y_seq = task.gen_seq(teseqlen,pm_trial_position=pmL)
#     x_seq,y_seq = task.gen_seq(teseqlen,None)
    x_embeds,ytarget = task.embed_seq(x_seq,y_seq)
    # forward prop
    yhat = net(x_embeds)
    ep_score = (ytarget == tr.softmax(yhat,-1).argmax(-1)).float().squeeze()
    score[ep] = ep_score 
  return score
```

```python
## sweeping params
seed = 0
tr.manual_seed(seed)
np.random.seed(seed)
```

```python
# task
pmweight = 1
pmtrials = 0
nback=1
ntokens_og=3

# input embeddings
signal = 8
edim = 16
noise = edim-signal
og_signal_dim = signal
pm_signal_dim = signal
og_noise_dim = noise
pm_noise_dim = noise

# network
indim = og_signal_dim+og_noise_dim
batch=1
outdim=3
```

```python
# model and task
EM = 0
stsize = 30
pdim = 9
net = PMNet(indim,pdim,stsize,outdim,EM)
task = NBackPMTask(nback,pmtrials,og_signal_dim,pm_signal_dim,og_noise_dim,pm_noise_dim,ntokens_og,seed)

# specify loss and optimizer
loss_weight = tr.FloatTensor([1,1,pmweight]) 
lossop = tr.nn.CrossEntropyLoss()
optiop = tr.optim.Adam(net.parameters(), lr=0.0005)

# ### train
print('train')

# training
nepochs = 10000
thresh = .9
trseqlen=20

acc = 0
nembeds = 0
L = []
for ep in range(nepochs):
  if ep%100==0:
    score = eval_(net,task,10,pmL=None)
    acc = score.mean()
    if acc>thresh:
      print('DONE1')
      break
  if ep%500==0: 
    print(ep/nepochs,acc)
  # generate data
  x_seq,y_seq = task.gen_seq(trseqlen)
  x_embeds,ytarget = task.embed_seq(x_seq,y_seq)
  # forward prop
  yhat = net(x_embeds)
  # collect loss through time
  loss = 0
  for yh,yt in zip(yhat,ytarget):
    loss += loss_weight[yt]*lossop(yh,yt)
  # bp and update
  optiop.zero_grad()
  loss.backward()
  optiop.step()

```
