### eval acc

```python
dataL = []
for seed in range(5):
  fpath = fdir+"LSTM25-EM1-nback_2-pm_maps_3-seed_%i-evacc.npy"%(seed)
  try:
    evacc = np.load(fpath)
    dataL.append(evacc)
  except:
    pass
  
evacc = np.array(dataL)
evacc.shape
```

```python
plt.figure(figsize=(13,4))
plt.axhline(0.33,c='lime')
plt.ylim(0,1.05)

M = evacc.mean(0)
S = evacc.std(0)/np.sqrt(len(evacc))

plt.plot(M,c='k',lw=3)
plt.scatter(range(len(M)),M,c='k')
for subacc in evacc:
  plt.plot(subacc,lw=1,c='b')
```

### evacc pmprobes over trials

```python
plt.figure(figsize=(13,4))
plt.axhline(0.33,c='lime')
plt.ylim(0,1.05)

pm_probe_idx = np.arange(6,130,13)
plt.bar(range(10),M[pm_probe_idx],yerr=S[pm_probe_idx])
plt.title("pm accuracy over trials")
```

### evacc by probe

```python
dataL = []
for seed in range(5):
  fpath = fdir+"LSTM25-EM1-nback_2-pm_maps_3-seed_%i-evacc_byprobe.npy"%(seed)
  try:
    evacc = np.load(fpath)
    dataL.append(evacc)
  except:
    pass
  
evacc2 = np.array(dataL)
evacc2.shape
```

# lstm states

### eval and save

```python
fdir = 'model_data/DualPM_sweep-2trials/'
```

```python
for seed in range(5):
  # load net
  fpath = fdir+"LSTM25-EM1-nback_2-pm_maps_3-seed_%i-model.pt"%(seed)
  net = NetDualPM(emsetting=1,seed=seed)
  try:
    net.load_state_dict(tr.load(fpath))
  except:
    continue  
  # eval
  acc,states = eval_net(net,task,neps=500,num_trials=5,trial_len=15,return_states=True)
  # save states
  fpath = fdir+"states/LSTM25-EM1-nback_2-pm_maps_3-seed_%i-states"%(seed)
  np.save(fpath,states)
```

### dim reduce visualize

```python
seed = 1
fpath = fdir+"states/LSTM25-EM1-nback_2-pm_maps_3-seed_%i-states.npy"%(seed)
states = np.load(fpath)

hstates = states[:,0,:,:]
cstates = states[:,1,:,:]
```

```python
from sklearn.decomposition import PCA
trial_len = 13


fig,axarr = plt.subplots(5,5,figsize=(20,20));axarr=axarr.reshape(-1)
for e in range(25):
  ax = axarr[e]
  pca = PCA(n_components=2)
  st_pc = pca.fit_transform(cstates[e,:,:])
  ax.scatter(st_pc[:trial_len,0],st_pc[:trial_len,1],c='r')
  ax.scatter(st_pc[trial_len:trial_len*2,0],st_pc[trial_len:trial_len*2,1],c='b')

title='PCA_cstates-first_two_trials'
plt.suptitle(title,y=0.9,fontsize=20)
plt.savefig('figures/DualPM_sweep-2trials/'+title)
```

### RSA

```python
from scipy.spatial import distance
compute_rdm = lambda M: distance.cdist(M,M,metric='euclidean')
```

```python
exp_len = 90
sub_rdms = -np.ones([500,exp_len,exp_len])
for i in range(500):
  sub_rdms[i] = compute_rdm(cstates[i])
```

```python
plt.imshow(sub_rdms.mean(0))
plt.savefig('figures/DualPM_sweep-2trials/'+'mean_RDM')
```
