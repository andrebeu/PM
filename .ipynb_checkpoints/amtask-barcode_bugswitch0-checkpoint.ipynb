# train acc


```python
# final accuracy
accL = []
xlabL = []
for wmsize,nmaps,switch,ntr in itertools.product(wmsizeL,nmapsL,switchL,ntrL):
  trsc = load_trsc(wmsize,nmaps,switch,ntr,nnets=20)
  final_trac = trsc[:,-500:,:,nmaps:].mean((1,2,3))
  accL.append(final_trac)
  xlabL.append('%i\n%i\n%i\n%i'%(wmsize,nmaps,switch,ntr))
# means
MaccL = [acc.mean() for acc in accL]
SaccL = [acc.std()/np.sqrt(len(acc)) for acc in accL]
xlab = xlabL
# plot
```

```python
plt.figure(figsize=(20,3))
ax = plt.gca()
ax.bar(range(len(MaccL)),MaccL,yerr=SaccL)
ax.set_ylim(.5,1)
ax.set_xticklabels(xlab,fontsize=20)
ax.set_xticks(range(len(MaccL)))
ax.text(-1.6,.43,'wmsize',fontsize=20)
ax.text(-1.6,.36,'nmaps',fontsize=20)
ax.text(-1.6,.3,'switch',fontsize=20)
ax.text(-1.6,.24,'ntrials',fontsize=20)
```

```python
def load_trsc_subj(wmsize,nmaps,switch,ntr,seed):
  ''' [neps,ntrs,nmaps+trlen]
  '''
  fdir = 'model_data/amtask-barcode2/'
  fpath = 'wmsize_%i-nmaps_%i-switch_%i-ntrials_%i-seed_%i.npy'%(wmsize,nmaps,switch,ntr,seed)
  return np.load(fdir+fpath)

def load_trsc(wmsize,nmaps,switch,ntr,nnets=20):
  ''' returns [nsubs,neps,ntrs,nmaps+trlen]'''
  L = []
  for seed in range(nnets):
    try:
      trac_subj = load_trsc_subj(wmsize,nmaps,switch,ntr,seed)
      L.append(trac_subj)
    except:
      continue
  return np.array(L)

def load_trac_subj(wmsize,nmaps,switch,ntr,seed):
  ''' mean over ts'''
  trsc = load_trsc_subj(wmsize,nmaps,switch,ntr,seed)
  trac = mov_avg(trsc[:,trial,nmaps:],1000)
  return trac
  
```

```python
def plt_group(ax,wmsize,nmaps,switch,ntr,nnets=20):

  MA = mov_avg
  trial_num = 1
  nsubs_plt=5

  trsc = load_trsc(wmsize,nmaps,switch,ntr,nnets=20)
  trac = trsc[:,:,trial_num,nmaps:].mean(-1)
  M = trac.mean(0)
  
  ax.plot(MA(M,100),c='b')
  for s in range(nsubs_plt):
    ax.plot(MA(trac[s],1000),lw=.3,c='k',zorder=0)
  return None


for wmsize,nmaps,switch,ntr in itertools.product(wmsizeL,nmapsL,switchL,ntrL):
  plt.close('all')
  ax = plt.gca()
  title = 'wmsize_%i-nmaps_%i-switch_%i-ntr_%i'%(wmsize,nmaps,switch,ntr)
  ax.set_title(title)
  plt_group(ax,wmsize,nmaps,switch,ntr,nnets=20)
  plt.savefig('figures/amtask-barcodes2/trac/group-'+title)
  
```

```python
''' plot subjects '''
nnets_plt = 20

cL = ['r','g','b']
for wmsize,nmaps,switch,ntr in itertools.product(wmsizeL,nmapsL,switchL,ntrL):
  plt.close('all')
  fig,axarr = plt.subplots(5,4,figsize=(20,12),sharex=True,sharey=True);axarr = axarr.reshape(-1)

  for trial in range(ntr):
    for sub in range(nnets_plt):
      ax = axarr[sub]
      ax.set_title(sub)
      trsc = load_trsc_subj(wmsize,nmaps,switch,ntr,sub)
      trac = mov_avg(trsc[:,trial,nmaps:],1000)
      ax.plot(trac,c=cL[trial])
      ax.set_ylim(.3,1.05)

  fname = 'wmsize_%i-nmaps_%i-ntr_%i-switch_%i'%(wmsize,nmaps,ntr,switch)
  plt.savefig('figures/amtask-barcodes2/trac/'+fname)
```

```python
emsetting=1
wmsize=4
instdim=10
stimdim=12
seed=2

net = NetBarCode(
        wmsize=wmsize,
        emsetting=emsetting,
        seed=seed,
        instdim=instdim,
        stimdim=stimdim,
        debug=False)

net.retrieve_mode = 'argmin'
net.emk_weights = tr.nn.Parameter([1,0]) # stim,lstm

nmaps = 4
switch = 1

task = TaskArbitraryMaps(
          nmaps,
          switchmaps=switch,
          ntokens_surplus=0,
          seed=seed,
          stimdim=stimdim)
```
