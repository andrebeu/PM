```python
def load_em_data():
  '''load presaved eval states and acc'''
  ntrials,seqlen = 6,5
  fpath = 'model_data/buckets_eval_data_N100/ntrials_%i-seqlen_%i'%(ntrials,seqlen)
  acc = np.load(fpath+'-acc.npy')
  states = np.load(fpath+'-states.npy')

  _,nnets,_ = acc.shape
  _,_,neps,_,_,_ = states.shape
#   print(states.shape) # (switch,nnets,eps,(h,c),tsteps,stsize)
#   print(acc.shape) # (switch,nnets,tsteps)

  '''reshape to remove instruction phase, select c or h state, and pull trials dimension'''
  # acc
  acc = acc.reshape(2,nnets,ntrials,10+seqlen)[:,:,:,10:]
  acc0,acc1 = acc

  # state
  state_class = 'hstate'
  states = states[:,:,:,int(state_class=='cstate'),:,:] # switch,nnets,neps,(h,c),tsteps,stsize
  states = states.reshape(2,nnets,neps,ntrials,10+seqlen,25)[:,:,:,:,10:,:]

  '''rename for comparison with lstms'''
  print(states.shape,'(switch,nnets,neps,ntrials,seqlen,stsize)') # (switch,nnets,neps,ntrials,seqlen,stsize)
  print(acc.shape,'(switch,nnets,ntrials,seqlen)') # (switch,nnets,ntrials,seqlen)
  return states,acc
```

## states RSA

```python
from scipy.spatial import distance
dist_metric = 'cosine'
compute_rdm = lambda M: distance.cdist(M,M,metric=dist_metric)

def get_mean_sub_rdm(states):
  ''' loop over epochs '''
  neps,tsteps,stsize = states.shape
  rdm = np.zeros([tsteps,tsteps])
  for ep in range(neps):
    rdm += compute_rdm(states[ep])
  rdm = rdm/neps
  return rdm

def compute_rdms(states):
  """ 
  loop over nets and epochs 
  given cstates [nnets,neps,tsteps,stsize]
  computes TxT rsm for each subj (averaged over neps)
  returns rsm per net [nnets,tsteps,tsteps]
  """
  # init 
  nnets,neps,tsteps,stsize = states.shape
  rsms = -np.ones([nnets,tsteps,tsteps])
  # loop compute RSM
  for sub in range(nnets):
    sub_rsms = -np.ones([neps,tsteps,tsteps])
    sub_state = states[sub]
    for ep in range(neps):
      sub_rsms[ep] = compute_rdm(sub_state[ep])
    rsms[sub] = sub_rsms.mean(0)
  return rsms
```

```python
def get_rdms_full(states):
  ''' takes rdsm [nnets,neps,ntrials,seqlen,stsize]
  returns [nnets,ntrials*seqlen,ntrials*seqlen]'''
  nnets,neps,ntrials,seqlen,stsize = states.shape
  states_full = np.concatenate([states[:,:,i,:,:] for i in range(ntrials)],-2)
  return compute_rdms(states_full)

```

```python
rdms_em0 = get_rdms_full(states_em[0])
rdms_em1 = get_rdms_full(states_em[1])

plt.figure(figsize=(15,15))
plt.title('RDM - switch1 - 6 blocks - 5 trials/block')
plt_rdm(plt.gca(),rdms_em1.mean(0),annot=True)
# plt.savefig('figures/buckets_eval_data_N100/meanRDM-switch1')
```

```python
''' comapring two trials '''

def get_rdms_tr0tr1(states,tr0,tr1):
  ''' takes rdsm [nnets,neps,ntrials,seqlen,stsize]
  returns [nnets,2*seqlen,2*seqlen]'''
  states_tr0tr1 = np.concatenate([states[:,:,tr0],states[:,:,tr1]],axis=-2)
  rdms = compute_rdms(states_tr0tr1)
  return rdms

def compute_separation(states,tr0,tr1):
  ''' 
  takes rdsm [nnets,neps,ntrials,seqlen,stsize]
  returns '''
  rdms_tr0tr1 = get_rdms_tr0tr1(states,tr0,tr1)
  sep = rdms_tr0tr1[:,seqlen:,:seqlen].mean((1,2))
  return sep
```

```python
def plt_rdm(ax,rdm,annot=False,plt_range=None):
  if type(plt_range)==type(None):
    plt_range = [rdm.min(),rdm.max()]
  sns.heatmap(rdm.round(2),ax=ax,
    vmin=plt_range[0],vmax=plt_range[1],
    cbar=False,annot=annot,center=0)
  return None
  
```

### comparison of trial separation

```python
def compute_sep_loop(states):
  ''' loop over tr0 and tr1
  takes states [nnets,neps,ntrials,trlen,stsize]
  returns [nnets,ntrials,ntrials]
  '''
  nnets,neps,ntrials,trlen,stsize = states.shape
  sep = np.ones([nnets,ntrials,ntrials])
  for tr0,tr1 in itertools.product(range(ntrials),range(ntrials)):
    sep[:,tr0,tr1] = compute_separation(states,tr0,tr1)
  return sep
  
```

```python
'''mean separation between states of two trials'''
sep_em0 = compute_sep_loop(states_em[0])
sep_em1 = compute_sep_loop(states_em[1])
sep_lstm = compute_sep_loop(states_lstm[1])
```

```python
##
sep0,sep1 = sep_lstm,sep_em1
sep0_lab,sep1_lab = 'lstm_only','switch'
##

ntrials_ = 6
fig,axarr = plt.subplots(ntrials_,ntrials_,figsize=(18,18),sharex=True,sharey=True)
plt.suptitle('representational distance between trials - x:%s, y:%s'%(sep0_lab,sep1_lab),y=.92,fontsize=20)
axarr[-1,0].set_xlabel(sep0_lab)
axarr[-1,0].set_ylabel(sep1_lab)

for tr0,tr1 in itertools.product(range(ntrials_),range(ntrials_)):
  ax = axarr[tr0,tr1]

  ax.plot(np.linspace(0,1,2),c='k',lw=.5)
  ax.scatter(sep0[:,tr0,tr1],sep1[:,tr0,tr1])
  
  sep_ratio = sum(sep1[:,tr0,tr1]/sep0[:,tr0,tr1] > 1)/len(sep_em0)
  ax.set_title('trials %i & %i\nratio = %.2f'%(tr0,tr1,sep_ratio),y=.87)
  ax.set_ylim(0,.8)
  ax.set_xlim(0,.8) 

plt.savefig('figures/buckets_eval_data_N100/separation_between_trials-%s_%s'%(sep0_lab,sep1_lab))
```

### relating separation and performance

```python
'''how much of the performance benefit of EM is related to induced separation'''

# difference in separation
sep_diff_tr0tr1 = sep_em1[:,0,1]-sep_em0[:,0,1]
# difference in overall accuracy
acc_diff = acc_em[1].mean((1,2)) - acc_em[0].mean((1,2))


plt.figure(figsize=(10,5))
ax = plt.gca()
rho,pval = pearsonr(acc_diff,sep_diff_tr0tr1)
ax.set_title('r=%.2f, p=%.4f'%(rho,pval))

sns.regplot(acc_diff,sep_diff_tr0tr1,ax=ax)
ax.set_xlabel('accuracy difference')
ax.set_ylabel('separation difference')
plt.savefig('figures/buckets_eval_data_N100/acc_diff-sep_diff-switch_noswitch')
```
