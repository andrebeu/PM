### eval acc analysis

```python
def plt_group_acc(ax,acc,c='blue',lab=None):
  """ acc [Nnets,time]
  """
  M = acc.mean(0)
  S = acc.std(0)/np.sqrt(len(acc))
  for subacc in acc:
    plt.plot(subacc,lw=.1,c=c)
  ax.plot(M,c=c,label=lab)
  ax.scatter(range(len(M)),M,c=c)
  ax.set_ylim(0,1.01)
  ax.set_ylabel('eval accuracy')
  return None
```

## cell state analysis

```python
from scipy.spatial import distance
compute_rdm = lambda M: distance.cdist(M,M,metric='euclidean')

def compute_rdms(cstates):
  """ 
  given cstates [nnets,neps,tsteps,stsize]
  computes TxT rsm for each subj (averaged over neps)
  """
  # init 
  nnets,neps,tsteps,stsize = cstates.shape
  rsms = -np.ones([nnets,tsteps,tsteps])

  # loop compute RSM
  for sub in range(nnets):
    sub_rsms = -np.ones([neps,tsteps,tsteps])
    sub_cstate = cstates[sub]
    for ep in range(neps):
      sub_rsms[ep] = compute_rdm(sub_cstate[ep])
    rsms[sub] = sub_rsms.mean(0)
  return rsms
```

```python
def plt_tr_boundary(ax):
  tsteps = ntrials*(ntokens+seqlen)
  for t in range(0,tsteps,ntokens+seqlen):
    ax.axvline(t,c='red',lw=.5)
    ax.axhline(t,c='red',lw=.5)
    ax.axvline(t+ntokens,c='purple',lw=.5)
    ax.axhline(t+ntokens,c='purple',lw=.5)
```

#### testphase RDMs

```python
def load_testphase_rdms(trial_num1,trial_num2,trep,switch,seqlen,ntrials,embool,ntokens=10,ev_seqlen=5):
  """
  specify trial_num1, trial_num2 to get the rsm between these two trials
    if trial_num1==trial_num2, this corresponds to within trial similarity
    if trial_num1!=trial_num2, between trial similarity
  return the rsm of the testphase of that trial
  """ 
  fpath = "LSTM25-EMcc-ntokens10-trep_%i-tr_switch_%i-tr_seqlen_%i-tr_ntrials_%i-ev_EM_%i-cstates.npy"%(
                      trep,switch,seqlen,ntrials,embool)
  cstates = np.load("model_data/buckets_eval_data/cstates/"+fpath)
  full_rsms = compute_rdms(cstates)

  # begin and end tsteps of test phases of desired trials
  trial_len = ntokens+ev_seqlen
  tr1_t0,tr1_t1 = (trial_num1*trial_len)+ntokens,(trial_num1+1)*trial_len
  tr2_t0,tr2_t1 = (trial_num2*trial_len)+ntokens,(trial_num2+1)*trial_len
  # crop rsm
  rdm = full_rsms[:,tr1_t0:tr1_t1,tr2_t0:tr2_t1]
  return rdm
```

```python
## compute testphase rdms and save

seqlenL = [5,13]
ntrialsL = [2,4]
switchL = [0,1]
trep = 9
embool = 1

for seqlen,ntrials,switch in product(seqlenL,ntrialsL,switchL):
  print('seqlen,ntrials,switch',seqlen,ntrials,switch)
  test_phase_rdm = -np.ones([10,11,11]) # sub,trial,trial
  # confirm that file exists
  try: 
    load_testphase_rdms(0,0,trep,switch,seqlen,ntrials,embool,ntokens=10)
  except:
    print('FAILURE:',seqlen,ntrials,switch)
    continue
  # loop for every pair of trial
  for tr1,tr2 in product(np.arange(11),np.arange(11)):
    print(tr1,tr2)
    trial_rdm = load_testphase_rdms(tr1,tr2,trep,switch,seqlen,ntrials,embool,ntokens=10)
    test_phase_rdm[:,tr1,tr2] = trial_rdm.mean((1,2))
  # save
  
  np.save('model_data/buckets_eval_data/testphase_rdms/'+fpath+'-testphase_means',test_phase_rdm)
  plt.imshow(test_phase_rdm)
  

```
